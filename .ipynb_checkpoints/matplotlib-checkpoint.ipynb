## matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
```

```python
x
```

## 直线图调用plot

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
cosy=np.cos(x)
cosy.shape
```

```python
siny=y.copy()
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle="--")
plt.show()
```

```python
siny=y.copy()
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle="--")
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
siny=y.copy()
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle="--")
plt.axis([-1,11,-2,2])

plt.show()
```

```python
siny=y.copy()
plt.plot(x,siny)
plt.plot(x,cosy,color='red',linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
siny=y.copy()
plt.plot(x,siny,label='sin(x)')
plt.plot(x,cosy,color='red',linestyle="--",label='cos(x)')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
siny=y.copy()
plt.plot(x,siny,label='sin(x)')
plt.plot(x,cosy,color='red',linestyle="--",label='cos(x)')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("welcome to the matplotlib world")
plt.show()
```

## 散点图调用Scatter Plot

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color='red')
plt.show()
```

```python
x=np.random.normal(0,1,100)
y=np.random.normal(0,1,100)
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.normal(0,1,1000)
y=np.random.normal(0,1,1000)
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.normal(0,1,1000)
y=np.random.normal(0,1,1000)
plt.scatter(x,y,alpha=0.5)
plt.show()
```
